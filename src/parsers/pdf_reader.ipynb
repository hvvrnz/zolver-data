```python
import fitz

doc = fitz.open("your.pdf")

targets = ["전공필수", "전공선택", "교양", "이수구분"]

for i in range(doc.page_count):
    text = doc.load_page(i).get_text()
    if any(t in text for t in targets):
        print(f"\n=== {i+1} 페이지 ===")
        print(text)
```
